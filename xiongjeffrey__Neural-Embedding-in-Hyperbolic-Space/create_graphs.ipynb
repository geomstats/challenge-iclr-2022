```python
import pickle
import numpy as np
from matplotlib import pyplot as plt

filenames=['largest_group_strings', 'strings_test', 'strings_subset','clean_strings']
for filename in filenames:
    (model,avg_loss,e_size,dist_types)=pickle.load(open('./models/'+filename+'.pkl', 'rb'))
    print(avg_loss)
    for i in range(len(dist_types)):
        plt.plot(e_size,avg_loss[:,i],label=dist_types[i])
    plt.legend()
    plt.xscale("log")
    xlabel('number of dimensions')
    ylabel('test loss')
    plt.title(filename)
    plt.show()
```
